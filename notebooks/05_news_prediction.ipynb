```python
import requests
import json
import pandas as pd
import re
import joblib
import os
import sys

from datetime import datetime
from bs4 import BeautifulSoup

sys.path.append('../')
from src.utilities.config_ import train_data_path, scrape_data_path, model_path
```

```python
# Load TFIDF
svm_file = "svm_model.pkl"
loaded_svm_model = joblib.load(os.path.join(model_path, svm_file))

# Load the saved TF-IDF vectorizer
vectorizer_file = "tfidf_vectorizer.pkl"
loaded_vectorizer = joblib.load(os.path.join(model_path, vectorizer_file))
```

```python
# read csv
dailyfx = pd.read_csv(os.path.join(scrape_data_path, "dailyforex_test.csv"))
econtimes = pd.read_csv(os.path.join(scrape_data_path, "economictimes_test.csv"))
```

```python
# New text data to predict
titles = list(dailyfx['title'])

# Preprocess the new text data
new_texts_tfidf = loaded_vectorizer.transform(titles)

# Predict the labels for the new text data
predictions = loaded_svm_model.predict(new_texts_tfidf)
dailyfx["label"] = predictions
dailyfx
```

```python
# New text data to predict
titles = list(econtimes['title'])

# Preprocess the new text data
new_texts_tfidf = loaded_vectorizer.transform(titles)

# Predict the labels for the new text data
predictions = loaded_svm_model.predict(new_texts_tfidf)
econtimes["label"] = predictions
econtimes
```
